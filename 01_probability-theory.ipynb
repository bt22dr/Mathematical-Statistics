- 결정모형(deterministic model) vs. 확률모형(probability model)
- 확률모형의 용어
 - 실험(experiment): 어떤 현상의 관찰 결과를 얻기 위한 과정
 - 표본공간(sample space): 모든 관찰 가능한 결과의 집합. 
   - S = {HH, HT, TH, TT}
 - 사건(event): 표본공간의 부분집합
   - A = {HT, TH}
- 확률의 정의의 해석
 - 고전적 확률정의: 
   - 표본공간이 유한개이고 모든 실험결과들이 등확률이라는 가정이 맞지 않는 경우 확률정의가 불가능
 - 상대도수를 이용한 확률정의: 실험을 무한히 반복하여 상대도수의 극한값은 확률로 정의함
   - 어떤 사건에 대해서는 무한반복시행이 불가능
 - 주관적 확률정의: 확률을 불확실성의 측도로 보는 접근방법. 베이지안 통계
 - 공리적 확률정의: 콜모고로프는 확률을 공리적으로 정의
   - https://en.wikipedia.org/wiki/Probability_axioms
   - 위 3가지 정의에서도 이 공리를 반드시 만족해야 함
- 조건부 확률
 - $ P(A\mid B)=\frac{P(A\cap B)}{P(B)} $
 - 전확률공식: $P(A)=\sum_{i=1}^k P(B_i)P(A\mid B_i)$
 - 베이즈 정리: $P(B_j\mid A)=\frac{P(B_j)P(A\mid B_j)}{\sum_{i=1}^k P(B_i)P(A\mid B_i)}$
 - 독립(independent): $P(A\cap B)=P(A)\cdot P(B)$
 - 서로 독립(mutually independent): $P(A_{i_1} \cap A_{i_2} \cap \cdots A_{i_k}) = P(A_{i_1}) \cdot P(A_{i_2}) \cdots P(A_{i_k})$

## 1.4 경우의 수

**곱의 법칙**: 사건 A가 일어나는 방법이 m가지이고, 그 각각에 대하여 사건 B가 일어나는 방법이 n가지이면, 사건 A와 B가 동시에 일어나는 방법은 $m\times n$가지이다. 이러한 내용을 곱의 법칙(multiplication rule)이라고 한다. 

**순열**: 서로 다른 n개의 원소 중에서 r개를 선택하여 '순서 있게(ordered)' 늘어놓는 것을 n개에서 r개 택한 순열(permutation)이라고 하고, 늘어놓는 경우의 수를 $_nP_r$로 표현한다. 곱의 법칙에 의해

$$
_nP_r = n(n-1)(n-2) \times \cdots \times (n-r+1) = \frac{n!}{(n-r)!}
$$

로 계산된다. 

**조합**: 서로 다른 n개의 원소 중에서 '순서에 관계없이' r개를 선택할 때 이 r개로 이루어진 각각의 집합을 n개에서 r개 택한 조합(combination)이라고 하고, 선택하는 조합의 수를 $\binom{n}{r}$ 또는 $_nC_r$로 표현한다. 그리고 이는 

$$
_nC_r = \frac{_nP_r}{r!} = \frac{n!}{r!(n-r)!}
$$

로 계산된다. 

**이항계수**(binomial coefficient): 다음의 이항전개식으로부터 

$$
(a+b)^n=\sum_{k=0}^{n}\binom{n}{k}a^kb^{n-k}
$$

이항계수 $\binom{n}{k}$는 $a^k b^{n−k}$  의 계수가 됨을 알 수 있다. (참고: https://www.mathfactory.net/11101)

이것을 k(>2) 종류의 원소가 있는 경우로 확장해보자. k개의 집단이 각각 $r_i\ (i = 1, 2, \cdots , k)$개의 한 종류 원소로 구성되어 있을 때, 전부 합친 $n(=r_1+r_2+ \cdots + r_k)$ 개의 원소를 늘어놓는 방법의 가짓수는

$$
\binom{n}{r_1 r_2 \cdots r_k} = \frac{n!}{r_1! r_2! \cdots r_k!}
$$

로 계산된다. 이를 이용하여 다항계수를 정의할 수 있다. 

**다항계수**(multinomial coefficient): 다음의 다항전개식으로부터 

$$
(a_1+a_2+...+a_k)^n=\sum_{r_1+...+r_k=n}^{ }\binom{n}{r_1r_2...r_k}a_1^{r_1}a_2^{r_2}...a_k^{r_k}
$$

다항계수 $\binom{n}{r_1 r_2…r_k}$는 $a_1^{r_1} a_2^{r_2}…a_k^{r_k}$의 계수가 됨을 알 수 있다. (참고: https://librewiki.net/wiki/%EB%8B%A4%ED%95%AD%EA%B3%84%EC%88%98)

**복원추출, 비복원추출**
주머니에 하얀 공이 5개, 검은 공이 10개 있다고 하자. 이 주머니에서 3개의 공을 무작위 추출할 때, 하얀공이 2개와 검은공 1개 뽑힐 확률을 구해보자. 
- 복원추출: 순서를 고려한 전체 가능한 가짓수는 $15^3$, 하얀공 2개와 검은공 1개 뽑하는 방법의 가짓수는 각각 $5^2$과 $10^1$이다. 그런데 WWB, WBW, BWW 모두 구하는 사건에 해당하므로 곱의 법칙에 의해 하얀 공이 2개, 검은 공이 1개 뽑히는 방법은 $3 \cdot 5^2 \cdot 10^1$이다. 따라서 구하고자 하는 확률은 
$$
\frac{3 \cdot 5^2 \cdot 10^1}{15^3}
$$
이 된다.

- 비복원추출: 전체 15개의 공에서 순서없이 3개의 공을 추출하는 방법은 $\binom{15}{3}$가지가 있으며, 하얀공이 2개 검은공이 1개 뽑히는 방법의 가짓수는 각각 $\binom{5}{2}$과 $\binom{10}{1}$이다. 곱의 법칙에 의해 하얀공 2개 검은공 1개 뽑히는 방법의 가짓수는 $\binom{5}{2}\binom{10}{1}$가지이며, 따라서 구하고자 하는 확률은
$$
\frac{\binom{5}{2}\binom{10}{1}}{\binom{15}{3}}
$$
이 된다.
