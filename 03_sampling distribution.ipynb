# 3. 표본분포

통계이론의 주된 활용은 자료로부터 연구대상의 성격을 규명하는 것이다. 전체 연구대상을 모두 조사하는 것은 대개 불가능하기 때문에 통계적 방법들은 그 일부인 표본을 조사함으로써 총체(totality)에 대한 성격을 귀납적으로 추론하는 길을 택한다. 이러한 귀납적 추론은 불확실성을 내포하고 있으므로 이를 계량화하기 위해 확률을 사용한다. 

이 장에서는 앞에서 다룬 확률분포이론을 통계이론에 적용하는 데에 있어 중요한 다리 역할을 하는 표본분포에 대해 다룰 것이다. 특히 정규분포로부터 추출한 랜덤표본에 근거한 함수의 분포로서
- 카이제곱분포
- F 분포
- t 분포

를 다룬다. 

# 3.1 모집단과 표본

**\[정의 3.1\]**  
연구 또는 관찰의 대상이 되는 총체(totality)를 모집단(population)이라 한다. 

**\[정의 3.2\]**  
같은 확률분포 $f(\cdot)$를 가지는, n개의 상호 독립(mutual independence)인 확률변수 $X_1, X_2, \cdots , X_n$이 $f_{X_1, \cdots , X_n}(x_1, x_2, \cdots , x_n)=f(x_1)f(x_2) \cdots f(x_n)$ 꼴의 결합 확률밀도함수를 가질 때 $X_1, X_2, \cdots , X_n$를 모확률밀도함수가 $f(\cdot)$인 크기가 n인 랜덤표본(random sample)이라고 한다. 또, 랜덤표본을 표현할 때 "확률변수 $X_1, X_2, \cdots , X_n$이 iid(independent and identically distributed)이다" 라고도 한다. 

이 정의에서 $X_i$는 확률변수를 의미하며, i번째 관찰변수가 가질 값을 뜻한다는 것이다. 일단 관찰된 알려진 값은 $x_1, x_2, \cdots , x_n$으로 소문자를 사용하여 표기한다.

**\[정의 3.3\]**  
미지의 모수를 포함하지 않는 랜덤표본의 함수 $T=T(X_1, X_2, \cdots , X_n)$를 통계량(statistic)이라고 한다. 통계량 자체도 확률변수이다. 여기서 모수 미포함의 의미는 다음과 같다. 모집단의 모수는 우리가 통계량으로부터 추정해야하는 값인데, 통계량에 모수가 포함되면 추정해야하는 모수의 함수가 되므로 순환논증에 빠져버린다. 그러므로 통계량에는 모수가 포함되면 안된다. 

### 3.1.1 표본적률

**표본적률의 정의와 성질**  
앞에서 다루었던 모분포에 대한 적률과 마찬가지로 표본에 대한 적률을 정의할 수 있다.

**\[정의 3.4\]**  
확률밀도함수 $f(x; \theta)$로부터 크기가 n인 랜덤표본 $X_1, X_2, \cdots , X_n$을 얻었다. 이때 r차 표본적률(sample moment)은

$$
{m_r}' = \frac{1}{n} \sum_{i=1}^{n} {X_i}^r
$$

으로, 표본평균 $\overline{X}_n = \sum_{i=1}^{n} X_i/n$에 대한 r차 표본중심적률(central sample moment)은 

$$
m_r = \frac{1}{n} \sum_{i=1}^{n} (X_i - \overline{X}_n)^r
$$

으로 정의된다. 

표본적률은 
- 표본적률은 확률표본의 함수이므로 통계량이다. 
- 모적률의 추정에 활용될 수 있다. (r차 표본적률 ${m_r}'$의 기대값은 r차 모적률 ${\mu_r}' = E(X^r)$이 되기 때문)
- 1차 표본적률은 표본평균, 2차 표본중심적률은 표본분산이다. 

**표본평균과 표본분산**  
모평균, 모분산의 추정에 활용되는 표본평균과 표본분산에 대해 알아보자. 먼저 표본평균과 표본분산의 기대값과 분산을 구해보자.

**\[정리 3.1\]**  
평균과 분산이 각각 $\mu$와 $\sigma^2$인 확률밀도함수 f(x)로부터 랜덤표본 $X_1, X_2, \cdots , X_n$을 얻었다. 이제 표본평균과 표본분산을 각각 

$$
\overline{X}_n = \sum_{i=1}^{n} \frac{X_i}{n} \qquad S^2 = \sum_{i=1}^{n} \frac{(X_i-\overline{X}_n)^2}{n-1}
$$

로 표기하면, 

$$
\begin{aligned}
& E(\overline{X}_n) = \mu \qquad  Var(\overline{X}_n) = {\sigma^2 \over n} \\
& E(S^2) = \sigma^2 \qquad        Var(S^2) = {1 \over n} \left ( \mu_4 - \frac{n-3}{n-1} \sigma^4 \right ) \qquad n>1
\end{aligned}
$$

이 성립한다. 위에서 표본분산을 구할 때 n이 아니라 n-1로 나누는 이유는 불편추정량으로 만들기 위함이다. Bessel's Correction([1](https://ghebook.blogspot.com/2010/09/statistics.html), [2](https://www.khanacademy.org/math/ap-statistics/summarizing-quantitative-data-ap/more-standard-deviation/v/review-and-intuition-why-we-divide-by-n-1-for-the-unbiased-sample-variance))를 참고하라.

증명  
1) $E(\overline{X}_n)$  
$$
\begin{aligned}
E(\overline{X}_n) & = E\left ( \sum_{i=1}^{n} {X_i \over n} \right ) \\
                  & = {1 \over n} E\left ( \sum_{i=1}^{n} X_i \right ) \\
                  & = {1 \over n} \sum_{i=1}^{n} E(X_i) \\
                  & = {1 \over n} n E(X) \\
                  & = \mu
\end{aligned}
$$

3번째 등식이 성립하는 이유는 [기대값의 성질](./02_probability-variables-and-distributions.ipynb#2.4.1-기대값과-분산의-성질) 때문이고 4번째 등식은 확률변수 $X_i$의 기대값이 모분포의 확률변수의 기대값 $E(X)$와 같기 때문이다. 

2) $Var(\overline{X}_n)$
$$
Var(\overline{X}_n) = \frac{1}{n^2} Var \left ( \sum_{i=1}^{n} X_i \right ) = \frac{1}{n^2} Var(X_i) = \frac{\sigma^2}{n}
$$

[분산의 성질](./02_probability-variables-and-distributions.ipynb#2.4.1-기대값과-분산의-성질)에 의해 위 과정이 성립한다. [다른 방식](https://www.youtube.com/watch?v=WfiRjHATlrg)으로 증명하는 방법도 있다. 

3) $E(S^2)$
표본분산의 기대값을 유도하기 위해 우선 $\sum_{i=1}^{n}(X_i-\mu)^2$를 구해보자. 

$$
\begin{aligned}
\sum_{i=1}^{n}(X_i-\mu)^2 & = \sum_{i=1}^{n} (X_i - \overline{X}_n + \overline{X}_n - \mu)^2 \\
                          & = \sum_{i=1}^{n} [(X_i-\overline{X}_n)^2 + 2(X_i-\overline{X}_n)(\overline{X}_n-\mu) + (\overline{X}_n-\mu)^2] \\
                          & = \sum_{i=1}^{n} [(X_i-\overline{X}_n)^2] + n(\overline{X}_n-\mu)^2 \qquad (\because \text{모든 편차의 합은 0이므로 }2(\overline{X}_n-\mu)\sum_{i=1}^{n}(X_i-\overline{X}_n)=0)
\end{aligned}
$$

이므로 

$$
\sum_{i=1}^{n} [(X_i-\overline{X}_n)^2] = \sum_{i=1}^{n}(X_i-\mu)^2 - n(\overline{X}_n-\mu)^2
$$

가 성립함을 알 수 있다. 따라서 표본분산의 기대값은

$$
\begin{aligned}
E(S^2) & = E \left [\sum_{i=1}^{n} \frac{(X_i-\overline{X}_n)^2}{n-1} \right ] \\
       & = \frac{1}{n-1} E \left [\sum_{i=1}^{n} (X_i-\overline{X}_n)^2 \right ] \\
       & = \frac{1}{n-1} E \left [ \sum_{i=1}^{n}(X_i-\mu)^2 - n(\overline{X}_n-\mu)^2 \right ] \qquad (\because \text{앞에서 구한 등식 사용})\\
       & = \frac{1}{n-1} \left [ \sum_{i=1}^{n}E (X_i-\mu)^2 - nE(\overline{X}_n-\mu)^2 \right ] \\
       & = \frac{1}{n-1} \left [ \sum_{i=1}^{n} {\sigma_{X_i}}^2 - n{\sigma_{\overline{X}_n}}^2 \right ] \\
       & = \frac{1}{n-1} \left [ n\sigma^2 - n\frac{\sigma^2}{n} \right ] \\
       & = \sigma^2
\end{aligned}
$$

이 성립한다. 5~7 번째 등식이 성립하는 이유는 다음과 같다. 첫 번째 항을 풀어쓰면 $E(X_1-\mu)^2+E(X_2-\mu)^2+\cdots E(X_n-\mu)^2$가 되며 각각은 표본개수가 n=1인 표본분산($\frac{\sigma^2}{n}=\frac{\sigma^2}{1}=\sigma^2$)으로 생각할 수 있다([참고](https://www.youtube.com/watch?v=mUnKM-XAA7g)). 그러므로 $\sum_{i=1}^{n}{\sigma_{X_i}}^2 = n\sigma^2$이다. 그냥 간단히 각 표본이 모집단을 따르므로 ${\sigma_{X_i}}^2 = \sigma^2$라고 생각해도 무방하다. 두 번째 항의 의미를 잘 살펴보면 표본평균($\overline{X}_n$)에서 표본평균의 기대값($E(\overline{X}_n)=\mu$)을 뺀 값의 제곱의 기대값이며 이는 표본평균의 분산이다. 표본평균의 분산은 ${\sigma_{\overline{X}_n}}^2 = {\sigma^2 \over n}$이다. 

이 증명 과정에서 추가로 알 수 있는 것은 표본분산을 구할 때 n-1이 아니라 n으로 나누어 주었다면 표본분산의 기대값$E(S^2)$이 $\sigma^2$이 아닌 $\frac{n-1}{n}\sigma^2$으로 구해졌을 것이라는 점이다.

**\[정의 3.5\]**  
확률변수의 열 $X_1, X_2, \cdots , X_n, \cdots$과 확률변수 $X$가 같은 확률공간에 정의된다고 하자. 만약 임의의 $\epsilon>0$에 대하여 $\lim_{n \rightarrow \infty} P(|X_n - X| \geq \epsilon)=0$, 또는 동등하게 $\lim_{n \rightarrow \infty} P(|X_n - X| < \epsilon)=1$을 만족한다면, $X_n$이 $X$로 확률적으로 수렴한다고 말하고, $X_n \overset{p}{\rightarrow} X$로 표기한다. 

**\[정의 3.6\]**  
확률변수의 열 $X_1, X_2, \cdots , X_n, \cdots$이 누적확률분포함수 $F_{X_1}, F_{X_2}, \cdots , F_{X_n}, \cdots$을 각각 갖고, 확률변수 X가 누적확률분포함수 $F_X$를 갖는다고 하자. 만약 함수 $F_X$가 연속인 모든 점 x에서 $\lim_{n->\infty}F_{X_n}(x) = F_X(x)$가 만족된다면, $X_n$이 X로 분포수렴한다고 말하고, $X_n \overset{d}{\rightarrow} X$로 표기한다. 

확률수렴은 확률변수에 대한 수렴을, 분포수렴은 PDF나 CDF에 대한 수렴을 말하며, 이 둘의 차이는 모수 포함 여부로 볼 수 있다. 뒤에서 살펴보겠지만 확률수렴의 대표적인 예는 대수의 법칙이고, 분포수렴의 대표적인 예는 중심극한정리이다. 

**\[정리 3.2\]**  
대수의 법칙 (weak) law of large numbers:  
평균이 $\mu<\infty$인 확률밀도함수 f(x)로부터 랜덤표본 $X_1, X_2, \cdots , X_n$을 얻었다면, n이 무한히 커질 때 $\overline{X}_n \overset{p}{\rightarrow} \mu$가 성립한다. 즉, 표본평균은 모평균으로 수렴한다는 의미이다. 

증명: 편의를 위해 확률밀도함수 f(x)가 분산 $\sigma^2<\infty$를 갖는다고 가정하자. 그러면,

$$
\begin{aligned}
P[|\overline{X}_n-\mu| < \epsilon] & = P[(\overline{X}_n-\mu)^2 < \epsilon^2] \\
                                   & = P[(\overline{X}_n-\mu)^2 < (\epsilon^2 \frac{n}{\sigma^2})(\frac{\sigma^2}{n})] \qquad (\because \text{체비세프 부등식과 } E(\overline{X}_n) = \mu , Var(\overline{X}_n) = \frac{\sigma^2}{n}) \\
                                   & \geq 1 - \frac{1}{\epsilon^2 \frac{n}{\sigma^2}} \\
                                   & = 1 - \frac{\sigma^2 / n}{\epsilon^2} \rightarrow 1
\end{aligned}
$$

임을 알 수 있다. 그러므로 정의 3.5에 의해 $\overline{X}_n \overset{p}{\rightarrow} \mu$가 성립한다. 

### 3.1.2 중심극한정리

**\[정리 3.3\]**  
중심극한정리: 평균과 분산이 각각 $\mu$와 $\sigma^2<\infty$인 확률밀도함수 f(x)로부터 랜덤표본 $X_1, X_2, \cdots , X_n$을 얻었다. 이때 확률변량

$$
Z_n = \frac{\sum_{i=1}^{n} X_i - E \left ( \sum_{i=1}^{n} X_i \right )}{\sqrt{Var\left ( \sum_{i=1}^{n} X_i \right )}} = \frac{\sum_{i=1}^{n} (X_i - \mu)}{\sqrt{n}\sigma} \\
(\because Var(\sum X_i) = Var(n \sum \frac{X_i}{n}) = n^2 Var(\sum \frac{X_i}{n}) = n^2 Var(\overline{X}_n) = n^2\cdot \frac{\sigma^2}{n} = n\sigma^2)
$$

의 분포는 표본의 크기 n이 무한대에 접근함에 따라 표준정규분포 N(0,1)로 수렴한다. 

중심극한정리 참고: [증명 1](https://blog.naver.com/psggoma/220899911971), [증명 2](http://freshrimpsushi.tistory.com/43), [동영상 재생목록](https://www.youtube.com/playlist?list=PLmljWRabIwWBxh8V6eIODIz--B802mdLt), [의의](http://www.ssacstat.com/base/cs/cs_05.php?com_board_basic=read_form&com_board_idx=401&topmenu=5&left=5&&com_board_search_code=&com_board_search_value1=&com_board_search_value2=&com_board_page=13&)

중심극한정리에서 가장 중요한 점 중의 하나는 모분포에 대해 특정한 꼴을 필요로 하지 않는다는 점이고, 모분포 형태에 관계없이 유한한 평균과 분산만 존재하면 확률변량 $Z_n$의 분포는 표준정규분포로 수렴한다는 사실이다. 

또한, 확률변량 $Z_n$은 

$$
Z_n = \frac{\overline{X}_n - E(\overline{X}_n)}{\sqrt{Var(\overline{X}_n)}} = \frac{\overline{X}_n - \mu}{\sigma / \sqrt{n}}
$$

와 같이 표본평균 $\overline{X}_n$를 사용해 표현할 수도 있는데, 이로부터 중심극한정리의 응용은 
 - 표본의 합 $\sum_{i=1}^{n} X_i$ 또는
 - 표본평균 $\overline{X}_n$
 
의 분포를 그에 상응하는 정규분포로 근사하는 데에 있다는 것을 알 수 있다. 

## 3.2 정규분포로부터의 표본추출

정규분포로부터 얻은 랜덤표본의 함수로 표현되는 확률변수들에 대해 살펴보자.

**\[정리 3.4\]**  
서로 독립인 확률변수 $X_i (i=1, \cdots , n)$들이 정규분포 $N(\mu_i, \sigma_i^2)$을 따르면, 그들의 합 $\sum_{i=1}^{n} X_i$는 정규분포 $N \left ( \sum_{i=1}^{n}\mu_i, \sum_{i=1}^{n}\sigma_i^2 \right )$을 따른다. 

증명: $\sum_{i=1}^{n}X_i$의 적률생성함수는

$$
\begin{aligned}
M(t) & = M_{X_1}(t)M_{X_2}(t) \cdots M_{X_n}(t) \\
     & = \prod_{i=1}^{n} e^{\mu_i t + \sigma_i^2t^2 / 2} \\
     & = exp \left [ t\sum_{i=1}^{n} \mu_i + {t^2 \over 2} \sum_{i=1}^{n} \sigma_i^2 \right ]
\end{aligned}
$$

이 되는데, 이는 바로 $N \left ( \sum_{i=1}^{n}\mu_i, \sum_{i=1}^{n}\sigma_i^2 \right )$의 적률생성함수이므로 정리 3.4가 성립한다는 것을 알 수 있다. 이는 앞에서 적률생성함수를 다룰 때 [정규분포확률변수의 합은 정규분포를 따른다](./02_probability-variables-and-distributions.ipynb#2.8.2-적률생성함수의-성질)는 것을 보이면서 증명한 바 있다.

사족으로 위 정리 3.4로부터, 평균이 $\mu$이고 분산이 $\sigma^2$인 정규분포로부터 구한 크기가 n인 랜덤표본의 표본평균을 $\overline{X}_n$라 하면, 아래와 같이 $\overline{X}_n \sim N(\mu, \sigma^2/n)$임을 알 수 있다. 

$$
\begin{aligned}
\overline{X}_n = \sum_{i=1}^{n} \frac{X_i}{n} & \sim N({1 \over n} \sum_{i=1}^{n} \mu, {1 \over n} \sum_{i=1}^{n} \sigma^2)  \qquad (\because E(aX) = aE(X), Var(aX) = a^2Var(X)) \\
                                              & \sim N(\mu, \frac{\sigma^2}{n})
\end{aligned}
$$

### 3.2.1 카이제곱분포
카이제곱분포는 모분산 $\sigma^2$의 추정량으로 사용되는 표본분산 $S^2=\sum_{i=1}^{n}\frac{(X_i-\overline{X_n})^2}{(n-1)}$의 표본분포를 이해하는데 중요한 역할을 한다. 

모수가 $(k, \theta)$인 감마분포에서 $k={n \over 2}, \theta=2$인 확률변수 X, 즉 $X \sim GAM({n \over 2},2)$는 자유도가 n인 카이제곱분포를 따른다고 한다. 

** \[정리 3.5\]**  
자유도가 n인 카이제곱분포의 확률밀도함수(PDF), 적률생성함수(MGF)는 아래와 같고,

$$
f_X(x) = \frac{1}{\Gamma({n \over 2})} \left ( {1 \over 2} \right )^{\frac{n}{2}} x^{{n\over2}-1} e^{-{x\over2}} \qquad x>0 \\
M_X(t) = (1-2t)^{-\frac{n}{2}} \qquad (t < {1 \over 2})
$$

기대값, 분산은 아래와 같다.

$$
E(X) = n, \quad Var(X) = 2n
$$

** \[정리 3.6\]**  
서로 독립인 확률변수 $X_i(i=1,\cdots ,n)$들이 각각 자유도가 $k_i$인 카이제곱분포를 따르면 그들의 합인 $Y=\sum_{i=1}^{n}X_i$는 자유도가 $\sum_{i=1}^{n}k_i$인 카이제곱분포를 따른다.

** \[정리 3.7\]**  
확률변수 $Z$가 $N(0,1)$ 분포를 따를 때, $Y=Z^2$은 $\chi^2(1)$ 분포를 가진다.

** \[정리 3.8\]**  
서로 독립인 확률변수 $X_i(i=1,\cdots,k)$들이 각각 정규분포 $N(\mu_i, \sigma_i^2)$을 따른다고 하면

$$
V=\sum_{i=1}^{k} \left ( \frac{X_i - \mu_i}{\sigma_i} \right )^2
$$

은 자유도가 k인 카이제곱분포를 따른다. 

** \[정리 3.9\]**  
$X_1, X_2, \cdots , X_n$을 평균이 $\mu$이고 분산이 $\sigma^2$인 정규분포 $N(\mu, \sigma^2)$으로부터의 크기가 n인 랜덤표본이라고 하면,

(1) $\overline{X_n}$와 $S^2=\sum_{i=1}^{n} \frac{(X_i-\overline{X_n})^2}{(n-1)}$은 서로 독립이며, 

(2) $(n-1)\frac{S^2}{\sigma^2}$은 자유도가 n-1인 카이제곱분포를 따른다. 

### 3.2.2 F 분포

F 분포는 정규분포로부터 구한 독립인 두 표본의 분산비에 대한 분포를 설명하는데 중요한 역할을 하며 분산분석에서 필수적으로 활용된다. 

먼저 확률변수를 살펴보자. 서로 독립인 카이제곱확률변수 U와 V의 자유도가 각각 n, m이라고 할 때, 그들의 함수꼴인 

$$
X = \frac{U/n}{V/m}
$$

은 자유도가 (n, m)인 F 분포를 따르며, X의 확률밀도함수는 

$$
f_X(x) = \frac{\Gamma[\frac{(n+m)}{2}]}{\Gamma(\frac{n}{2})\Gamma(\frac{m}{2})} \left ( \frac{n}{m} \right )^{{n \over 2}} \frac{x^{\frac{(n-2)}{2}}}{[1+\frac{nx}{m}]^{\frac{(n+m)}{2}}} \qquad (x>0)
$$

으로 구해진다. 위와 같은 확률밀도함수를 갖는 변수 X를 자유도가 (n, m)인 F 확률변수라고 하며, $F \sim F(n, m)$으로 표기한다. 

** \[정리 3.10\]**  
$X \sim F(n, m)$일 때 X의 평균과 분산은 아래와 같다. 

$$
\begin{aligned}
& E(X) = \frac{m}{m-2} \qquad & m>2\\
& Var(X) = \frac{2m^2(m+n-2)}{n(m-2)^2(m-4)} \qquad & m>4
\end{aligned}
$$

** \[정리 3.11\]**  
평균이 $\mu_X$이고 분산이 $\sigma_X^2$인 정규분포로부터 크기가 n인 랜덤표본 $X_1, X_2, \cdots , X_n$ 그리고 평균이 $\mu_Y$이고 분산이 $\sigma_Y^2$인 정규분포로부터 크기가 m인 랜덤표본 $Y_1, Y_2, \cdots , Y_n$을 얻었다. 이제 두 표본이 서로 독립이라고 하면 확률변량

$$
F = \frac{S_X^2/\sigma_X^2}{S_Y^2/\sigma_Y^2}
$$

은 자유도가 (n-1, m-1)인 F 분포를 따른다. (정리 3.9에서 표본분산과 모분산의 비율은 카이제곱분포를 따른다는 것을 보았으므로, 카이스제곱확률변수의 비율이 F 분포를 따른다는 의미가 된다)

### 3.2.2 t 분포

t 분포는 정규분포의 모평균에 대한 가설검정 등을 포함한 검정론에서 매우 중요한 역할을 한다. 

t 분포의 확률밀도함수는 다음과 같다.

$$
F_X(x) = \frac{\Gamma(\frac{k+1}{2})}{\Gamma(\frac{k}{2})} \frac{1}{\sqrt{k\pi}} \frac{1}{(1+\frac{x^2}{k})^{\frac{k+1}{2}}} \qquad (-\infty < x < \infty)
$$

이러한 확률밀도함수를 갖는 변수 X를 자유도가 k인 t 확률변수라고 하며, $X \sim t(k)$로 표기한다. 

t 확률변수의 분포 형태는 0에 대하여 대칭이며 모수인 자유도 k에 의해 확률밀도함수의 꼬리부분의 두터움이 결정된다. 자유도가 커지면 표준정규분포에 가까워진다. 

**\[정리 3.12\]**  
평균이 $\mu$이고 분산이 $\sigma^2$인 정규분포로부터의 랜덤표본 $X_1, X_2, \cdots , X_n$을 고려하면, 확률변량

$$
T = \frac{\sqrt{n}(\overline{X_n}-\mu)}{\sqrt{\frac{\sum_{i=1}^{n}(X_i - \overline{X_n})^2}{n-1}}}
$$

은 자유도가 (n-1)인 t 분포를 따른다. 

**\[정리 3.13\]**  
$X \sim t(k)$일 때 X의 평균과 분산은 아래와 같이 주어진다. 

$$
E(X) = 0, \quad Var(X)=\frac{n}{n-2} \qquad (n>2)
$$
